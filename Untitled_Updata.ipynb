## Data Preprocessing:

### We Notice that there are 12,121 empty celles(NaN value), so we remove them

```python
comments = comments.dropna()
```

### preprocess the comments text by lemmatize, removing stop words, and other necessary cleaning steps. 

```python
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))
REPLACE_BY_SPACE_RE = re.compile('[/{}()\[\]\|@,;]')
custom_meaningless_words = ['none','nul','nun']

def clean_text(text):
    text = text.lower()
    text= re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+?]|[!*\(\),]|'\
                 '(?:%[0-9a-fA-F][0-9a-fA-F]))+','',text)
    text = re.sub(REPLACE_BY_SPACE_RE," ",text)
    text = BAD_SYMBOLS_RE.sub('', text)
    lemmatizer = WordNetLemmatizer()
    text =' '.join(lemmatizer.lemmatize(word) for word in text.split(' ') \
                   if word.isalpha() and word not in STOPWORDS and word not in custom_meaningless_words and word in words )

    return text
```

```python
cleaned_comments = [clean_text(text) for text in comments['ParticipantResponse']]
cleaned_comments = [text for text in cleaned_comments if len(text)>2]
```

## Tokenize Data

```python
tokenizer = nltk.tokenize.word_tokenize
cleaned_tokens=[]
for text in cleaned_comments:
    tokens = tokenizer(text)
    tokens = [token for token in tokens if len(token)>2]
    cleaned_tokens.append(tokens)
```

## we use the VADER sentiment analysis tool from the nltk library 

```python
nltk.download('vader_lexicon')
def sentiment_analyzer(comment):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_scores = analyzer.polarity_scores(comment)

    if sentiment_scores['compound'] >= 0.05:
        sentiment = "Positive"
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    return comment, sentiment
```

```python
coms = []
sents = []
for comment in cleaned_comments:
    com, sent = sentiment_analyzer(comment)
    coms.append(com)
    sents.append(sent)
```

```python
df = pd.DataFrame({'comment': coms, 'sentiment': sents})
```

## Topic Modeling:

### We apply topic modeling techniques Latent Dirichlet Allocation (LDA) to identify common themes or topics within the feedback. This will help us group feedback items that discuss similar subjects.

### Step1: We will apply on negative comments only

```python
neg = list(df['comment'][df['sentiment'] == 'Negative'].index)
neg_list = []
for i in neg:
    neg_list.append(cleaned_tokens[i])
```

```python
from gensim import corpora
from gensim.models import LdaModel
dictionary = corpora.Dictionary(neg_list)
corpus = [dictionary.doc2bow(text) for text in neg_list]
```

```python
from gensim.models.coherencemodel import CoherenceModel
chunksize = 400
passes = 20 
iterations = 500
eval_every = None
num_topics_list = [5, 10, 15, 20, 25]

# Loop through different values and evaluate coherence scores
for num_topics in num_topics_list:
    lda_model = LdaModel(corpus=corpus, id2word=dictionary, chunksize=chunksize, \
                       alpha='auto', eta='auto', \
                       iterations=iterations, num_topics=num_topics, \
                       passes=passes, eval_every=eval_every,random_state=100)
    coherence_model = CoherenceModel(model=lda_model, texts=neg_list, dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model.get_coherence()
    print(f"Num Topics: {num_topics}, Coherence Score: {coherence_score}")
```

```python
num_topics =5
%time model = LdaModel(corpus=corpus, id2word=dictionary, chunksize=chunksize, \
                       alpha='auto', eta='auto', \
                       iterations=iterations, num_topics=num_topics, \
                       passes=passes, eval_every=eval_every,random_state=100)
```

```python
for topic_id, topic_words in model.print_topics():
    print(topic_id + 1,[x for x in topic_words.split('\"') if x.isalpha()][:5])
```

# Insights and Recommendations:

### about Negative comments

Topic 1: Teaching Style (Initial Experiences):

Insights: This topic focuses on initial experiences, which can significantly impact students' perceptions of a course. Interactions with instructors during the first week can set the tone for the entire semester.
Potential Actions:
Conduct instructor training or orientation to emphasize the importance of a positive first impression.
Encourage instructors to establish clear communication and expectations from the beginning.
Implement a mentorship program to provide support for new instructors.

Topic 2: Discomfort with Teaching (Course Format and Effectiveness):

Insights: Discomfort and disagreement with the course format or teaching methods can hinder the learning experience.
Potential Actions:
Gather detailed feedback from students to pinpoint specific aspects of discomfort.
Review and update course materials or instructional approaches based on feedback.
Provide resources for instructors to enhance their teaching effectiveness.

Topic 3: Independence in Learning (Desire for Autonomy):

Insights: Students express a desire for more independence in their learning journey, indicating a preference for self-directed learning.
Potential Actions:
Explore the integration of self-paced learning modules or resources.
Offer guidance on self-regulated learning strategies.
Encourage students to take ownership of their learning through goal setting.

Topic 4: Discomfort with Teaching (Similar to Topic 2):

Insights: This topic reiterates concerns related to teaching methods or materials, reinforcing the importance of addressing these issues.
Potential Actions:
Continuously collect feedback and conduct assessments to identify specific areas of discomfort.
Collaborate with instructors to redesign course materials or adjust instructional strategies.
Provide professional development opportunities for instructors.
Topic 5: Course Difficulty (Challenges in Learning):

Insights: Feedback indicates that students perceive the course as challenging, potentially leading to difficulties in understanding and managing workload.
Potential Actions:
Review course syllabi and assignments to ensure they align with learning objectives.
Offer additional support resources, such as tutoring or study groups.
Communicate expectations regarding the course's level of difficulty more clearly.

### about Positive comments

Areas for Improvement:

Ease of Use or Learning (Topic 2):

The presence of words like "need," "easier," and "power" suggests that students may find certain aspects of the courses challenging or less user-friendly. Improvements in course materials, tools, or instructions may be needed to enhance the learning experience.

Agreement on Issues (Topic 4): 

While agreement among students on certain issues can be a strength, it can also indicate areas of concern. It's important to investigate the nature of the issues that students are agreeing on and take action to address them.

Preferences in Learning (Topic 5): 

Understanding students' preferences in learning is valuable, but it's also essential to consider whether these preferences align with effective teaching practices. There may be opportunities to incorporate preferred teaching methods while maintaining educational effectiveness.

Potential Actions:

Enhance User-Friendly Learning: 

Based on feedback related to ease of use or learning (Topic 2), consider conducting a usability assessment of course materials and tools. Make improvements to ensure a more intuitive and user-friendly experience for students.

Engage with Students:

For areas where students are in agreement (Topic 4), engage in open communication with students to understand their concerns better. Conduct surveys or focus group discussions to gather more specific feedback and ideas for improvement.

Incorporate Preferred Learning Methods: 

Take into account students' preferences in learning (Topic 5) when designing and delivering courses. Experiment with different teaching styles and seek feedback from students to assess their effectiveness.

Celebrate Strengths: 

Acknowledge and celebrate areas of strength, such as positive feedback (Topic 1 and Topic 3). Recognize instructors and programs that receive praise and use these positive examples to inspire improvements in other areas.

Continuous Assessment and Improvement: 

Establish a process for ongoing assessment and improvement of courses. Regularly collect and analyze feedback from students, identify emerging topics and sentiments, and adapt course materials and teaching methods accordingly.

Support and Professional Development: 

Provide support and professional development opportunities for instructors to enhance their teaching skills and address areas that may need improvement.

Transparency and Communication: 

Maintain open and transparent communication channels with students. Share information about changes, improvements, and actions taken in response to feedback.
