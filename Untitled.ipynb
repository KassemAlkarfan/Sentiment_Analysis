## Data Preprocessing:

### preprocess the comments text by lemmatize, removing stop words, and other necessary cleaning steps. 

```python
comments = comments[comments.str.len() > 2]
```

```python
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))
REPLACE_BY_SPACE_RE = re.compile('[/{}()\[\]\|@,;]')
custom_meaningless_words = ['nun', 'hzhzhahz', 'hahshshha', 'ss', 'huhuhu', 'jygj', 'jhbkm', 'hhhh', 'hhh']

def cleaning_result(text):
    text = text.lower()
    text= re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+?]|[!*\(\),]|'\
                 '(?:%[0-9a-fA-F][0-9a-fA-F]))+','',text)
    text = re.sub(REPLACE_BY_SPACE_RE," ",text)
    text = BAD_SYMBOLS_RE.sub('', text)
    lemmatizer = WordNetLemmatizer()
    text =' '.join(lemmatizer.lemmatize(word) for word in text.split() if word.isalpha() and word not in STOPWORDS and word not in custom_meaningless_words ) 

    return text
```

```python
cleaned_tokens_list=[cleaning_result(text) for text in comments]
```

## we use the VADER sentiment analysis tool from the nltk library 

```python
nltk.download('vader_lexicon')
def sentiment_analyzer(comment):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_scores = analyzer.polarity_scores(comment)

    if sentiment_scores['compound'] >= 0.05:
        sentiment = "Positive"
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    return comment, sentiment
```

```python
coms = []
sents = []
```

```python
for comment in cleaned_tokens_list:
    com, sent = sentiment_analyzer(comment)
    coms.append(com)
    sents.append(sent)
```

```python
df = pd.DataFrame({'comment': coms, 'sentiment': sents})
```

## Topic Modeling:

### We apply topic modeling techniques Latent Dirichlet Allocation (LDA) to identify common themes or topics within the feedback. This will help us group feedback items that discuss similar subjects.

```python
tokenizer = nltk.tokenize.WhitespaceTokenizer()
cleaned_tokens=[tokenizer.tokenize(cleaning_result(text)) for text in cleaned_tokens_list]
```

```python
from gensim import corpora

dictionary = corpora.Dictionary(cleaned_tokens)
corpus = [dictionary.doc2bow(text) for text in cleaned_tokens]
```

```python
from gensim.models import LdaModel

num_topics = 5
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
for topic_id, topic_words in lda_model.print_topics():
    print(topic_id + 1,[x for x in topic_words.split('\"') if x.isalpha()])
```
